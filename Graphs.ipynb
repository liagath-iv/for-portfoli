```python
# DataFrame
#  (line chart),         2023 .
#  (bar chart),         2024 .
import pandas as pd
import matplotlib.pyplot as plt
def line_chart(x):
    data = pd.read_csv(x)
    df = pd.DataFrame(data)
    df=df.query(' == 2023')
    plt.plot(df[''], df[''],linestyle='-', marker = 'D', color = 'firebrick')
    plt.grid(True)
    plt.legend(' ')
    plt.xlabel('')
    plt.ylabel(' ')
    plt.title('  2023')
    return plt.show()
def bar_chart(x):
    data = pd.read_csv(x)
    df = pd.DataFrame(data)
    df=df.query(' == 2024')
    plt.bar(df[''], df[''], color = 'darkolivegreen')
    plt.grid(True)
    plt.legend(' ')
    plt.xlabel('')
    plt.ylabel(' ')
    plt.title('  2024')
    return plt.show()
x = 'books.csv'
# for 2023
line_chart(x)
#for 2024
bar_chart(x)
```
