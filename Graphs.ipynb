```python
#Загрузите DataFrame
#Линейный график (line chart), отображающий количество прочитанных книг по месяцам только за 2023 год.
#Столбчатую диаграмму (bar chart), отображающую количество прочитанных страниц по месяцам только за 2024 год.
import pandas as pd
import matplotlib.pyplot as plt
def line_chart(x):
    data = pd.read_csv(x)
    df = pd.DataFrame(data)
    df=df.query('Год == 2023')
    plt.plot(df['Месяц'], df['Книги'],linestyle='-', marker = 'D', color = 'firebrick')
    plt.grid(True)
    plt.legend('Прочитано книг')
    plt.xlabel('Месяцы')
    plt.ylabel('Количество книг')
    plt.title('Прочитанные книги 2023')
    return plt.show()
def bar_chart(x):
    data = pd.read_csv(x)
    df = pd.DataFrame(data)
    df=df.query('Год == 2024')
    plt.bar(df['Месяц'], df['Страницы'], color = 'darkolivegreen')
    plt.grid(True)
    plt.legend('Прочитано страниц')
    plt.xlabel('Месяцы')
    plt.ylabel('Количество страниц')
    plt.title('Прочитанные страницы 2024')
    return plt.show()
x = 'books.csv'
# for 2023
line_chart(x)
#for 2024
bar_chart(x)
```
