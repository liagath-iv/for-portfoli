```python
#Какое среднее количество просмотренных страниц пользователями сайта во всей таблице?
#Какое стандартное отклонение времени, проведенного на сайте? 
#Рассчитайте коэффициент вариации и сделайте вывод о том, насколько изменчиво поведение пользователей сайта в зависимости от времени нахождения на сайте. 
#Проверьте, есть ли статистически значимая разница в количестве страниц, просмотренных пользователями с мобильных устройств и десктопов.

import pandas as pd
import numpy as np
from scipy import stats
def website_analyse(s):
    df = pd.read_csv(s)
    df_extra_1 = df.query('`Тип устройства` == "Мобильное"')
    df_extra_2 = df.query('`Тип устройства` == "Десктоп"')
    pages = df["Просмотренные страницы"]
    pages_df1 = df_extra_1["Просмотренные страницы"]
    pages_df2 = df_extra_2["Просмотренные страницы"]
    time = df["Время на сайте (сек)"]
    k3 = np.std(time)/np.mean(time)
    stat, p_value = stats.ttest_ind(pages_df1, pages_df2)
    if k3<0.1:
        s1 = 'Слабая вариабельность'
    elif (k3>=0.1) and (k3<0.25):
        s1 = 'Умеренная вариабельность'
    else:
        s1 = 'Высокая вариабельность'
    if p_value > 0.05:
        s2 = 'Есть статистически значимая разница в количестве страниц'
    else:
        s2 = 'Нет статистически значимой разницы'
    return print(f'1) Среднее количество просмотренных таблиц: {np.mean(pages):.2f}\n2) Стандартное отклонение: {np.std(time):.2f}\n3) {s1}: {k3:.2f}\n4) {s2}: {p_value:.2f}')
s = 'website_visits_december_2024.csv'
website_analyse(s)
```
